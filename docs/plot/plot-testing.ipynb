```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from quantbullet.plot.core import LineStyler, PlotStyler, ScatterStyler
```

```python
# Generate a time series with 100 data points
time = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.sin(np.linspace(0, 10 * np.pi, 100)) + np.random.normal(0, 0.1, 100)
hue_values = np.sin(np.linspace(0, 10 * np.pi, 100)) + np.random.normal(0, 0.1, 100)

# Create a DataFrame
time_series_data = pd.DataFrame({'Date': time, 'Value': values, 'Hue': hue_values})
```

## Line Chart

```python
styler = (
    PlotStyler()
    .set_title("Time Series Data")
    .set_xlabel("Date")
    .set_ylabel("Value")
    .set_x_date_format( '%m %Y' )
    .set_y_percent()
)

linestyler = LineStyler()
```

```python
fig, ax = plt.subplots(figsize=(10, 5));
ax = styler.plot_line(
    ax,
    time_series_data["Date"],
    time_series_data["Value"],
    line_styler=linestyler.set_color("blue"),
)

ax = styler.plot_line(
    ax,
    time_series_data["Date"],
    time_series_data["Value"] + 0.5,
    line_styler=linestyler.set_color("red"),
)
```

## Scatter with Hue

```python
styler = (
    PlotStyler()
    .set_title("Time Series Data")
    .set_xlabel("Date")
    .set_ylabel("Value")
    .set_x_date_format( '%m %Y' )
    .set_y_percent()
)

linestyler = LineStyler()

fig, ax = plt.subplots(figsize=(10, 5));
ax = styler.plot_scatter_with_hue(
    ax,
    time_series_data["Date"],
    time_series_data["Value"],
    time_series_data["Hue"],
)
```

## Line and Scatter Chart

```python
styler = (
    PlotStyler()
    .set_title( "Time Series Data" )
    .set_xlabel( "Date" )
    .set_ylabel( "Value" )
    .set_x_date_format( '%m %Y' )
    .set_y_percent( True )
)

fig, ax = plt.subplots(figsize=(10, 5));
ax = styler.plot_line(
    ax,
    time_series_data["Date"],
    time_series_data["Value"],
    line_styler=linestyler.set_color("blue"),
)

ax = styler.plot_scatter(
    ax,
    time_series_data["Date"],
    time_series_data["Value"] + 0.5,
    scatter_styler = ScatterStyler().set_color("red"),
)
```
