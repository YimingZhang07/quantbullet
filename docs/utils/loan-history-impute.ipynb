```python
import pandas as pd
import numpy as np

# 原始数据，包括静态字段 original_amount
df = pd.DataFrame({
    'loan_id': [101, 102, 103],
    'n_periods': [12, 6, 3],
    'original_amount': [100000, 200000, 150000]
})
```

```python
# 计算每个贷款需要生成多少行（包括 period=0）
periods = df['n_periods'] + 1

# 用 repeat 展开每个字段
repeated_loan_ids = np.repeat(df['loan_id'].values, periods)
repeated_amounts = np.repeat(df['original_amount'].values, periods)
periods_each = np.concatenate([np.arange(p) for p in periods])

# 构建展开后的 DataFrame
expanded_df = pd.DataFrame({
    'loan_id': repeated_loan_ids,
    'period': periods_each,
    'original_amount': repeated_amounts
})
```
